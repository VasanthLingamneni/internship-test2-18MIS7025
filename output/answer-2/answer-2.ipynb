```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/VasanthLingamneni/internship-test2-18MIS7025/master/input/question-2/main.csv")
data.head()
```

```python
data.groupby('product_description')['price'].mean()
```

```python
data["price_new"] = data['price'].fillna(
   data.groupby('product_description')['price'].transform("mean")
)
```

```python
data["sales_amount"]=data["sales_quantity"].where(data["unit"] !="pcs",data["product_description"].str.split("-",expand=True)[1].astype("float") *\
data["sales_quantity"]
)
```
